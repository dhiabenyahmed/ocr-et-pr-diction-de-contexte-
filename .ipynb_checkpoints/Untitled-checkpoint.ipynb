```python
import os
os.environ['KMP_DUPLICATE_LIB_OK'] = 'TRUE'

import PyPDF2
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

import fitz  
import easyocr
import pandas as pd
import tensorflow as tf
import numpy as np
```

```python
# Fonction pour charger les donnes  partir de fichiers
def load_data(directory):
    texts = []
    labels = []
    for label in os.listdir(directory):
        class_folder = os.path.join(directory, label)
        if os.path.isdir(class_folder):
            for fname in os.listdir(class_folder):
                with open(os.path.join(class_folder, fname), 'r', encoding='utf-8') as file:
                    texts.append(file.read())
                    labels.append(label)
    return texts, labels
```

```python
# Fonction pour extraire le texte d'un fichier PDF
def extract_text_from_pdf(pdf_path, output_txt1):
    with open(pdf_path, 'rb') as pdf_file:
        pdf_reader = PyPDF2.PdfReader(pdf_file)
        
        with open(output_txt1, 'w', encoding='utf-8') as output_file:
            for page_num in range(len(pdf_reader.pages)):
                page = pdf_reader.pages[page_num]
                text_page = page.extract_text()
                output_file.write(text_page)
```

```python
# Charger les donnes d'entranement et de test
train_texts, train_labels = load_data('C:/Users/bdhia/OneDrive/Bureau/projet OCR/data/SANAD_SUBSET/khaleej/Train')
test_texts, test_labels = load_data('C:/Users/bdhia/OneDrive/Bureau/projet OCR/data/SANAD_SUBSET/khaleej/Train')
```

```python
# Encodez les tiquettes
label_encoder = LabelEncoder()
train_labels = label_encoder.fit_transform(train_labels)
test_labels = label_encoder.transform(test_labels)
```

```python
# Paramtres
vocab_size = 10000
embedding_dim = 16
max_length = 128
trunc_type = 'post'
padding_type = 'post'
oov_tok = "<OOV>"
```

```python
# Tokeniser et rembourrer les squences
tokenizer = Tokenizer(num_words=vocab_size, oov_token=oov_tok)
tokenizer.fit_on_texts(train_texts)
word_index = tokenizer.word_index

train_sequences = tokenizer.texts_to_sequences(train_texts)
train_padded = pad_sequences(train_sequences, maxlen=max_length, padding=padding_type, truncating=trunc_type)

test_sequences = tokenizer.texts_to_sequences(test_texts)
test_padded = pad_sequences(test_sequences, maxlen=max_length, padding=padding_type, truncating=trunc_type)
```

```python
# Construire le modle LSTM
model = Sequential([
    Embedding(vocab_size, embedding_dim, input_length=max_length),
    LSTM(64, return_sequences=True),
    LSTM(64),
    Dense(24, activation='relu'),
    Dense(len(label_encoder.classes_), activation='softmax')
])

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Entraner le modle
model.fit(train_padded, train_labels, epochs=10, validation_data=(test_padded, test_labels))
```

```python
# valuer le modle
loss, accuracy = model.evaluate(test_padded, test_labels)
print(f"Test Loss: {loss:.4f}, Test Accuracy: {accuracy * 100:.2f}%")
```

```python
# Enregistrez le modle aprs l'entranement
model.save('C:/Users/bdhia/OneDrive/Bureau/projet OCR/modele.h5')
```

```python
# Charger le modle  partir du fichier sauvegard
loaded_model = load_model('C:/Users/bdhia/OneDrive/Bureau/projet OCR/modele.h5')

# Chemin vers le fichier PDF
pdf_path ="C:/Users/bdhia/OneDrive/Bureau/projet OCR/politique.pdf"

# Chemin vers le fichier texte de sortie
output_txt_path = "C:/Users/bdhia/OneDrive/Bureau/projet OCR/test3.txt"

# Extraire le texte du PDF
extract_text_from_pdf(pdf_path, output_txt_path)

# Lire le texte extrait
with open(output_txt_path, 'r', encoding='utf-8') as txt_file:
    pdf_text = txt_file.read()

# Tokenize et rembourrer les squences
pdf_sequences = tokenizer.texts_to_sequences([pdf_text])
pdf_padded = pad_sequences(pdf_sequences, maxlen=max_length, padding=padding_type, truncating=trunc_type)

# Prdire la classe
prediction = loaded_model.predict(pdf_padded)
predicted_class = np.argmax(prediction)

# Convertir l'index prdit en la classe relle
predicted_class_name = label_encoder.inverse_transform([predicted_class])[0]

# Afficher les rsultats
print("Classe prdite par le modle de classification:", predicted_class_name)
```
